# 6.create a flag for orders here delivery_days is missing but is_delivered id true

```python
df["is_delivered_flag"] = (
    df["is_delivered"]
    .astype(str)
    .str.lower()
    .isin(["true", "yes", "1"])
)


df["missing"] = (
    df["delivery_days"].isna() & df["is_delivered_flag"]
)
df['missing']
```

# 7.create a flag column that marks orders where delivery days are missing, but the order is marked as delivered

```python
df['delivery_miss_but_delivered'] = (df['delivery_days'].isna() & df['is_delivered'].astype(str).str.lower().isin(['true','yes']))
df['delivery_miss_but_delivered'].value_counts()
```

# 9. Identify cities where more than 30% of orders have missing order_date.

```python
city_missing = (df.assign(missing = df['order_date'].isna())
                         .groupby('city')['missing']
                         .mean()
                         .mul(100)
)
city_missing[city_missing > 30]
```
