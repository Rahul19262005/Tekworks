## step:3 connect the database


```python
conn = connector.connect(
    host="localhost",
    user="root",
    password="123456",
    database="Student_db"
)
cursor = conn.cursor()
print("connection executed",conn.is_connected())
```

## step4: create table

```python
cursor.execute("CREATE TABLE IF NOT EXISTS Student (id INT PRIMARY KEY, name VARCHAR(255),age INT)")  #if we want to execute the query then execute is 
#required #"IF NOT EXIST" if it not present create if exist don't create
print("Table created successfully")
```

## step5: insert values (insert single,insert multiple)

```python
cursor.execute("INSERT INTO Student (id, name, age)VALUES (2, 'Rahul', 20)")

print("Record inserted successfully")
```
